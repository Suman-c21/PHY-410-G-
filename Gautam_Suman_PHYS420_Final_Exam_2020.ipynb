where $c_1 = 1.55\times 10^{-4} D$, $c_2 = 0.22~ D^2$, $D = 80.0 \times 10^{-6}$ m, $m = 2.68 \times 10^{-10}$ kg, and $\vec{v}$ is the velocity of the droplet.
```python
# Program to calculate the trajectory of droplet
# using midpoint  and Euler's Method
#
# This code was modified from the one discussed in class to 
# work in Python 3, as you will be using Jupyter Notebooks 
# for this exam instead of CASA.
#

#Load libraries
from scipy import *
import sys
from matplotlib.pyplot import *
from numpy import loadtxt
from pylab import *


# Code parameters
tau = 0.001                                       # Seconds. Delta_T, step in numerical code.

#constants 
D=80.0e-6                                         # Diameter of droplet in meter.
m=2.68e-10                                        # Mass in kg
C1=(1.55e-4)*D     
C2=0.22*D**2

# Initial conditions
speed_0 = 11.2                          # Initial speed in m/s.

vx_0 = speed_0                          # Horizontal component of speed in m/s.
vy_0 = 0.0                              # Verticle component of speed in m/s.

x_0 = 0.0                               # Initial horizontal position in meter.
y_0 = 1.5                               # Verticle position in meter.

# Acceleration
a_x = 0.0                               # Acceleration in m/s2 a_x
a_y = 9.8                               # Acceleration in m/s2 a_y


    
# Euler Method: trajectory

x_Euler = [0.0]
y_Euler = [0.0]
t_Euler = [0.0]
vx_Euler = [0.0]
vy_Euler = [0.0]

x_Euler[0] = x_0
y_Euler[0] = y_0
t_Euler[0] = 0.0
vx_Euler[0] = vx_0
vy_Euler[0] = vy_0

i = 0 # variable to count steps
y_current = y_Euler[i]

while (y_current >0):
    v=np.sqrt(vx_Euler[i]**2+vy_Euler[i]**2)
    a_x = -(C1+C2*v)*vx_Euler[i]/m                                      # Acceleration m/s2 a_x
    a_y = -9.8+(C1+C2*v)*vy_Euler[i]/m                                  # Acceleration m/s2 a_y
   
    
    x_Euler.append(x_Euler[i] + tau*vx_Euler[i])
    y_Euler.append(y_Euler[i] + tau*vy_Euler[i])

    vx_Euler.append(vx_Euler[i] + tau*a_x)
    vy_Euler.append(vy_Euler[i] + tau*a_y)

    i = i+1

    t_Euler.append(tau*i)

    y_current = y_Euler[i]


print('\n Maximum horizontal displacement by Euler method is: %1.2f meter. \n' % (x_Euler[i]))
print('\n Vertical displacement by Euler method is: %1.2f meter. \n' % (y_current))
print('\n Time of flight  by Euler method is: %1.2f second.  \n' % (t_Euler[i])) 



# Midpoint Method: trajectory

x_mid = [0.0]
y_mid = [0.0]
t_mid = [0.0]
vx_mid = [0.0]
vy_mid = [0.0]

x_mid[0] = x_0
y_mid[0] = y_0
t_mid[0] = 0.0
vx_mid[0] = vx_0
vy_mid[0] = vy_0


a_x = 0.0                                                  # Acceleration in m/s2 a_x
a_y = -9.8                                                 # Acceleration in m/s2 a_y

i = 0                                                       # variable to count steps
y_current = y_mid[i]

while (y_current >0):
    v=np.sqrt(vx_mid[i]**2+vy_mid[i]**2)
    a_x = 0.0-(C1+C2*v)*vx_mid[i]/m                        # Acceleration in m/s2 a_x
    a_y = -9.8+(C1+C2*v)*vy_mid[i]/m                       # Acceleration in m/s2 a_y
    
    
    
    vx_mid.append(vx_mid[i] + tau*a_x)
    vy_mid.append(vy_mid[i] + tau*a_y)
    
    x_mid.append(x_mid[i] + tau*0.5*(vx_mid[i]+vx_mid[i+1]))
    y_mid.append(y_mid[i] + tau*0.5*(vy_mid[i]+vy_mid[i+1]))

    i=i+1
    
    t_mid.append(tau*i)

    y_current = y_mid[i]


print('\n Maximum horizontal displacement by Miidpoint method is: %1.2f meter. \n' % (x_mid[i]))
print('\n Vertical displacement by Midpoint method is: %1.2f meter.\n' % (y_current))
print('\n Time of flight by Midpoint method is: %1.2f second.  \n' % (t_mid[i])) 


############## Plot data #################

clf() 

Euler_plot = plot(x_Euler,y_Euler,'s',markersize = 3,color='b', label = 'Euler Solution')
Midpoint_plot = plot(x_mid,y_mid,'s',markersize = 3,color='r', label = 'Midpoit method Solution')

xlabel(r'X(m)',fontsize=20)
ylabel(r'Y(m)',fontsize=20)
legend(numpoints=1)


show()

```
