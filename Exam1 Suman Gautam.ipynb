Vectors are fundamental in many areas of physics and engineering, and some vector identities are particularly useful. One example is the following triple product: 

$\vec{A} \cdot (\vec{B} \times \vec{C}) = (\vec{A} \times \vec{B}) \cdot \vec{C}$

```python
 '''This program ask the user for the three components of vectors and calculate the  cross product of two vectors
and finally, calculate the vector triple product  and verify the vector identity.'''

# 1. Asking the user for the cartesian components of three vectors a,b,c.

import sys                                                              # Import system.
import numpy as np                                                      # Import python Library.
a = np.zeros(4)                                                         # Initializing the vectors.
b = np.zeros(4)                                              
c = np.zeros(4)

axb = np.zeros(4)                                                       # Initializing the cross product of a and b.
bxc = np.zeros(4)                                                       # Initializing the cross product of b and c.


for i in range(1,4):                                                    # Using for loop.
    a[i]=float(input('Enter the component of vector A, a%d = ' %i))
for i in range(1,4):    
    b[i]=float(input('Enter the component of vector B, b%d = ' %i))
for i in range(1,4):
    c[i]=float(input('Enter the component of vector C, c%d = ' %i))
    
    
# 2.Calculating the cross product axb and bxc.         

axb[1]=a[2]*b[3]-a[3]*b[2]
axb[2]=a[3]*b[1]-a[1]*b[3]
axb[3]=a[1]*b[2]-a[2]*b[1]

                                                                       
bxc[1]=b[2]*c[3]-b[3]*c[2]
bxc[2]=b[3]*c[1]-b[1]*c[3]
bxc[3]=b[1]*c[2]-b[2]*c[1]


print('The cross product of a and b is:')                              # Printing the cross product of a and b. 
print(axb)


print('The cross product of b and c is:')                             # Printing the cross product of b and c.
print(bxc)



# 3. Calcualting the vector triple product (a.(bxc))
abxc = (a[1]*b[2]*c[3] - a[1]*b[3]*c[2] + a[2]*b[3]*c[1] - a[2]*b[1]*c[3] + a[3]*b[1]*c[2] - a[3]*b[2]*c[1])


# Calculating the vector triple product ((axb).c))
axbc = (a[2]*b[3]*c[1] - a[3]*b[2]*c[1] + a[3]*b[1]*c[2] - a[1]*b[3]*c[2] + a[1]*b[2]*c[3] - a[2]*b[1]*c[3])


print('The triple product of (a.bxc) is:')               # Printing the triple product of a dot b cross c.
print(abxc)

print('The triple product of (axb.c) is:')               # Printing the triple product of a cross b dot c.
print(axbc)


# 4. If statement to check the vector identity.

if (abxc==axbc):
    print('The two vector triple product a.bxc and axb.c are equal.')
else :
    print('The two vector triple product a.bxc and axb.c are not equal.')

```
